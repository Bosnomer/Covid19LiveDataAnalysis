```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('savedData/covid19data.csv')
data = data.iloc[:, 1:]
data
```

```python
Day = data[data['Current Status']=='Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()
Day
```

```python
x = np.arange(len(Day))
x
```

```python
y = Day.values
y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = x.reshape(-1, 1) #Converting to 2D data
x
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)                                        # y = b0 + b1*x   => Straight Line
```

```python
y[121]
```

```python
y[123]
```

```python
lr.predict([[124]])
```

```python
y
```

```python
Yr = lr.predict(x)
Yr
```

```python
plt.scatter(x, y)
plt.plot(x, Yr)
plt.show()
```

```python
lr.score(x, y)*100
```

```python
lr.intercept_                      #intercept value (Bo)
```

```python
lr.coef_                           #slope value (B1)
```
