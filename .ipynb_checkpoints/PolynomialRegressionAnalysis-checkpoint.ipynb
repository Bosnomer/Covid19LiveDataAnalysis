```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('savedData/covid19data.csv')
data = data.iloc[:, 1:]
data
```

```python
Day = data[data['Current Status']=='Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()
Day
```

```python
x = np.arange(len(Day))
x
```

```python
y = Day.values
y
```

```python
x = x.reshape(-1, 1)
x
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
X = pf.fit_transform(x)
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)                                     # y = b0 + b1*(x^0) + b2*(x^1) + b3*(x^2) 
```

```python
lr.coef_                                        # b1, b2, b3
```

```python
lr.intercept_                                   # b0
```

```python
Yp = lr.predict(X)
Yp
```

```python
plt.scatter(x, y)
plt.plot(x, Yp, color='k')
plt.show()
```

```python
lr.score(X, y)*100
```

```python
x
```

```python
pf.transform([[123]])
```

```python
lr.predict(pf.transform([[124]]))
```

# Now lets do the same regression with degree = 5

```python
Day = data[data['Current Status']=='Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()

x = np.arange(len(Day))
x = x.reshape(-1, 1)
y = Day.values
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=5)
X = pf.fit_transform(x)
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)                    # y = b0 + b1*(x^0) + b2*(x^1) + b3*(x^2) + b4*(x^3) + b5*(x^4) + b6*(x^5) 
```

```python
lr.coef_                                        # b1, b2, b3, b4, b5, b6
```

```python
lr.intercept_                                   # b0
```

```python
Yp = lr.predict(X)
Yp
```

```python
plt.scatter(x, y)
plt.plot(x, Yp, color='k')
plt.show()
```

```python
lr.score(X, y)*100
```

```python
lr.predict(pf.transform([[124]]))
```
